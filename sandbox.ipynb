```python
import numpy as np
import pandas as pd
import matplotlib as plt
```

```python
#read input files
consumptions = pd.read_csv('datasets\drinks.csv')
countries = pd.read_csv('datasets\countryContinent.csv', encoding = "ISO-8859-1")

#merge relevant data.
cols = ['country', 'continent', 'sub_region']

#using join function to join tables based on country name
c2 = consumptions.set_index('country').join(countries[cols].set_index('country'))
c2.head()
```
