(Optional) Write a function that returns all the column pairs with a correlation higher than 0.7 or lower than -0.7

Example output: 

- Wheelbase and Length: 0.89
- MPG_Highway and Weight: -0.79

```python
corr = cars.corr()

#this is slower >17ms
def correlationpairs(corr):
    for a, b in corr[(abs(corr[:])<1) & (abs(corr[:])>0.7)].items():
        b = b[b[:] > 0]
        list(map(lambda x: print(f'{a} and {x}: {corr[a][x]:.2f}'), b.index.values))

#faster implementation <4ms
def correlationpairs2(corr):
    for a, cols in corr.items():
        for c in cols.index.values:
            if((corr[a][c] > 0.7) and (corr[a][c]<1)):
                print(f'{a} and {c}: {corr[a][c]}')

    
```

```python
%timeit correlationpairs2(corr)
```

```python
%timeit correlationpairs(corr)
```
